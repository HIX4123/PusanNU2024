###f-strings을 이용한 문자열 포매팅(Formatting)

```python
FirstName = "재석"; LastName = "유"; job1 = "개그맨"; job2 = "MC"
print("저는 %s이자 %s인 %c%s 입니다." % ( job1, job2, LastName, FirstName))
```

```python
FirstName = "재석"; LastName = "유"; job1 = "개그맨"; job2 = "MC"
print(f"저는 {job1}이자 {job2}인 {LastName}{FirstName} 입니다.")
```

```python
name = "유재석"; weight = 67; height = 175.4
print(F"{name} 님의 체중은 {weight}kg이고, 키는 {height}cm입니다.")
```

```python
name = "유재석"; height = 175.4
print(F"{name:7s}님의 키는 {height:+7.1f}cm입니다.")
print(F"{name:^7s}님의 키는 {height:7.2f}cm입니다.")
print(F"{name:>7s}님의 키는 {height:07.0f}cm입니다.")
```
